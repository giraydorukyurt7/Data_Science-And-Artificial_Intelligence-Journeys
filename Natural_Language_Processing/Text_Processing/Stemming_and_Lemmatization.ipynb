```python
import nltk
from nltk.stem import PorterStemmer
from nltk.stem import WordNetLemmatizer
from nltk.corpus import wordnet
```

```python
some_tokens = ["walk","walking","walked","walks","going", "run","ran","running","bosses","was","were","is","are","am","replacement","natural","languages","processing","unneccessary", "berry", "mice","better"]
sentence = "lemmatization is more sophisticated than stemming".split()
sentence2 = "John Wick has a devoted following".split()
sentence3 = "The cat was following the bird as it flew by".split()
```

```python
porter = PorterStemmer()

for token in some_tokens:
    print(porter.stem(word=token))

print("\n"+"="*30+"\n")

for token in sentence:
    print(porter.stem(word=token), end=" ")
```

```python
lemmatizer = WordNetLemmatizer()

for token in some_tokens:
    print(lemmatizer.lemmatize(word=token))

print("\n"+"="*30+"\n")

for token in sentence:
    print(lemmatizer.lemmatize(word=token), end=" ")

print("\n"+"="*30+"\n")

for token in some_tokens:
    print(lemmatizer.lemmatize(word=token, pos=wordnet.VERB))

print("\n"+"="*30+"\n")

print(lemmatizer.lemmatize(word="going"))
print(lemmatizer.lemmatize(word="going", pos=wordnet.VERB))
print(lemmatizer.lemmatize(word="walking"))
print(lemmatizer.lemmatize(word="walking", pos=wordnet.VERB))
print(lemmatizer.lemmatize(word="better"))
print(lemmatizer.lemmatize(word="better", pos=wordnet.ADJ))
```

```python
#make this automatically
def get_wordnet_pos(treebank_tag):
    if treebank_tag.startswith("J"):
        return wordnet.ADJ
    elif treebank_tag.startswith("V"):
        return wordnet.VERB
    elif treebank_tag.startswith("N"):
        return wordnet.NOUN
    elif treebank_tag.startswith("R"):
        return wordnet.ADV
    else:
        return wordnet.NOUN
    
nltk.download('averaged_perceptron_tagger')
```

```python
print(sentence2)

words_and_tags = nltk.pos_tag(sentence2)
print(words_and_tags)

for word, tag in words_and_tags:
    lemma = lemmatizer.lemmatize(word=word, pos=get_wordnet_pos(tag))
    print(lemma, end=" ")
```

```python
print(sentence3)

words_and_tags = nltk.pos_tag(sentence3)
print(words_and_tags)

for word, tag in words_and_tags:
    lemma = lemmatizer.lemmatize(word=word, pos=get_wordnet_pos(tag))
    print(lemma, end=" ")
```
